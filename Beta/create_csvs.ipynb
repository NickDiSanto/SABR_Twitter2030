```python
# Creates CSVs with all counts

import csv

team_abbrevs = ['ANA', 'OAK', 'SEA', 'HOU', 'TEX', 'COL', 'SDP', 'LAD', 'SFG', 'AZ', 'STL', 'CIN', 'CHC', 'PIT', 'MIL',
                'ATL', 'NYM', 'PHI', 'FLA', 'WSN', 'CLE', 'MIN', 'KCR', 'DET', 'CHW', 'NYY', 'TBD', 'BOS', 'BAL', 'TOR']

header = ['Team', 'Date', 'WinLoss', 'HapSen', 'SadSen', 'GenSen', 'PosPol', 'NeuPol', 'NegPol', 'ComPol', 'YesCog', 'NoCog']

# Runs through each team + year
for team in team_abbrevs:
    win_loss = {}

    for i in range(16, 20):
        # Atlanta 2016 has bad results. Skip
        if team == 'ATL' and i == 16:
            continue
            
        # Gets all Win/Loss counts for each team+year
        with open(('Data/Team_WL/' + team + str(i) + '.csv'), newline='', encoding='utf-8-sig') as readfile:
            reader = csv.reader(readfile, delimiter=',', quotechar='|')
            for row in reader:
                date = row[0]
                wL = row[1]
                if (date.startswith('d')):
                    continue
                if not date.startswith('1'):
                    date = '0' + date
                if not date[5] == '/':
                    date = date[:3] + '0' + date[3:]
                date = date[6:] + '-' + date[:2] + '-' + date[3:-5]
                
                win_loss[date] = wL
                                            
    # Writes csv file of results
    # with open('Data/Output/' + team + '/IP/results.csv', 'w', newline='') as writefile:
    with open('Data/Output/' + team + '_IP_results.csv', 'w', newline='') as writefile:
        writer = csv.writer(writefile)
        writer.writerow(header)

        hapSen = {}
        # with open(('Data/Output/' + team + '/IP/Sentiment/happy.txt'), 'r') as readfile:
        with open(team + '_IP_sent_happy.txt', 'r') as readfile:
            for row in readfile:
                hapSen[row[:10]] = row[row.find(':')+2:]

        sadSen = {}
        # with open(('Data/Output/' + team + '/IP/Sentiment/sad.txt'), 'r') as readfile:
        with open(team + '_IP_sent_sad.txt', 'r') as readfile:
            for row in readfile:
                sadSen[row[:10]] = row[row.find(':')+2:]
        
        genSen = {}
        # with open(('Data/Output/' + team + '/IP/Sentiment/general.txt'), 'r') as readfile:
        with open(team + '_IP_sent_general.txt', 'r') as readfile:
            for row in readfile:
                genSen[row[:10]] = row[row.find(':')+2:]

        posPol = {}
        # with open(('Data/Output/' + team + '/IP/Polarity/positive.txt'), 'r') as readfile:
        with open(team + '_IP_pol_positive.txt', 'r') as readfile:
            for row in readfile:
                posPol[row[:10]] = row[row.find(':')+2:]

        neuPol = {}
        # with open(('Data/Output/' + team + '/IP/Polarity/neutral.txt'), 'r') as readfile:
        with open(team + '_IP_pol_neutral.txt', 'r') as readfile:
            for row in readfile:
                neuPol[row[:10]] = row[row.find(':')+2:]

        negPol = {}
        # with open(('Data/Output/' + team + '/IP/Polarity/negative.txt'), 'r') as readfile:
        with open(team + '_IP_pol_negative.txt', 'r') as readfile:
            for row in readfile:
                negPol[row[:10]] = row[row.find(':')+2:]

        comPol = {}
        # with open(('Data/Output/' + team + '/IP/Polarity/compound.txt'), 'r') as readfile:
        with open(team + '_IP_pol_compound.txt', 'r') as readfile:
            for row in readfile:
                comPol[row[:10]] = row[row.find(':')+2:]

        yesCog = {}
        # with open(('Data/Output/' + team + '/IP/Cognition/yes.txt'), 'r') as readfile:
        with open(team + '_IP_cogn_yes.txt', 'r') as readfile:
            for row in readfile:
                yesCog[row[:10]] = row[row.find(':')+2:]

        noCog = {}
        # with open(('Data/Output/' + team + '/IP/Cognition/no.txt'), 'r') as readfile:
        with open(team + '_IP_cogn_no.txt', 'r') as readfile:
            for row in readfile:
                noCog[row[:10]] = row[row.find(':')+2:]
                
        for date in win_loss.keys():
            if (date in hapSen.keys() and date in sadSen.keys() and date in genSen.keys() and date in posPol.keys() and date in neuPol.keys() and 
               date in negPol.keys() and date in comPol.keys() and date in yesCog.keys() and date in noCog.keys()):
                data = [team, date, win_loss[date], hapSen[date], sadSen[date], genSen[date], posPol[date], neuPol[date], negPol[date], comPol[date], yesCog[date], noCog[date]]

                writer.writerow(data)
            
print('El Fin')
```
