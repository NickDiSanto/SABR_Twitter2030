```python
import csv

# team_abbrevs = ['ANA', 'OAK', 'SEA', 'HOU', 'TEX', 'COL', 'SDP', 'LAD', 'SFG', 'AZ', 'STL', 'CIN', 'CHC', 'PIT', 'MIL', 'ATL', 'NYM', 'PHI', 
#                 'FLA', 'WSN', 'CLE', 'MIN', 'KCR', 'DET', 'CHW', 'NYY', 'TBD', 'BOS', 'BAL', 'TOR']
team_abbrevs = ['ANA']

header = ['Team', 'Date', 'WinLoss', 'HapSen', 'SadSen', 'GenSen', 'PosPol', 'NeuPol', 'NegPol', 'ComPol', 'YesCog', 'NoCog']

with open('Data/mlb_results.csv', 'w', newline='') as writefile:
    writer = csv.writer(writefile)
    writer.writerow(header)
    
    for team in team_abbrevs:
        
        winLoss = {}
        # for i in range(16, 20):
        for i in range(18, 19):
            with open(('Data/Team_WL/' + str(team) + str(i) + '.csv'), newline='', encoding='utf-8-sig') as readfile:
                reader = csv.reader(readfile, delimiter=',', quotechar='|')
                for row in reader:
                    date = row[0]
                    wL = row[1]
                    if (date.startswith('d')):
                        continue
                    if not date.startswith('1'):
                        date = '0' + date
                    if not date[5] == '/':
                        date = date[:3] + '0' + date[3:]
                    date = date[6:] + '-' + date[:2] + '-' + date[3:-5]

                    winLoss[date] = wL
                    
                    
        hapSen = {}
        # with open(('Data/Output/Sentiment/' + str(team) + '_happy.txt'), 'r') as readfile:
        with open(('Data/Output/Sentiment/ANA_happy.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    hapSen[row[:10]] = row[row.find(':')+2:]
                
        sadSen = {}
        # with open(('Data/Output/Sentiment/' + str(team) + '_sad.txt'), 'r') as readfile:
        with open(('Data/Output/Sentiment/ANA_sad.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    sadSen[row[:10]] = row[row.find(':')+2:]
                
        genSen = {}
        # with open(('Data/Output/Sentiment/' + str(team) + '_general.txt'), 'r') as readfile:
        with open(('Data/Output/Sentiment/ANA_general.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    genSen[row[:10]] = row[row.find(':')+2:]
                
        posPol = {}
        # with open(('Data/Output/Polarity/' + str(team) + '_positive.txt'), 'r') as readfile:
        with open(('Data/Output/Polarity/ANA_positive.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    posPol[row[:10]] = row[row.find(':')+2:]
                
        neuPol = {}
        # with open(('Data/Output/Polarity/' + str(team) + '_neutral.txt'), 'r') as readfile:
        with open(('Data/Output/Polarity/ANA_neutral.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    neuPol[row[:10]] = row[row.find(':')+2:]
                
        negPol = {}
        # with open(('Data/Output/Polarity/' + str(team) + '_negative.txt'), 'r') as readfile:
        with open(('Data/Output/Polarity/ANA_negative.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    negPol[row[:10]] = row[row.find(':')+2:]
                
        comPol = {}
        # with open(('Data/Output/Polarity/' + str(team) + '_compound.txt'), 'r') as readfile:
        with open(('Data/Output/Polarity/ANA_compound.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    comPol[row[:10]] = row[row.find(':')+2:]
                
        yesCog = {}
        # with open(('Data/Output/Cognition/' + str(team) + '_yes.txt'), 'r') as readfile:
        with open(('Data/Output/Cognition/ANA_yes.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    yesCog[row[:10]] = row[row.find(':')+2:]
                
        noCog = {}
        # with open(('Data/Output/Cognition/' + str(team) + '_no.txt'), 'r') as readfile:
        with open(('Data/Output/Cognition/ANA_no.txt'), 'r') as readfile:
            for row in readfile:
                if (row[:4] == '2018'):
                    noCog[row[:10]] = row[row.find(':')+2:]
                
        for date in winLoss.keys():
            if (date in hapSen.keys() and date in sadSen.keys() and date in genSen.keys() and date in posPol.keys() and date in neuPol.keys() and 
               date in negPol.keys() and date in comPol.keys() and date in yesCog.keys() and date in noCog.keys()):
                data = [team, date, winLoss[date], hapSen[date], sadSen[date], genSen[date], posPol[date], neuPol[date], negPol[date], comPol[date], yesCog[date], noCog[date]]

                writer.writerow(data)
            
print('El Fin')
```
